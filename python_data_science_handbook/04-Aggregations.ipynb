# Aggregations: Min, Max, and Everything In Between
- Using the NumPy built-in functions to summarize the "typical" values in a dataset.

```python
# Summing the Values in an Array
import numpy as np
L = np.random.random(100)
%timeit sum(L)
```

```python
# Using the NumPy Function.
%timeit np.sum(L)
```

```python
big_array = np.random.rand(1000000)
min(big_array), max(big_array)
```

```python
np.min(big_array), np.max(big_array)
```

```python
print(big_array.min(), big_array.max())
```

## Multi-dimensional Aggregates

```python
M = np.random.random((3,4))
print(M)
```

```python
M.sum()
```

```python
# min value in each column
print(M.min(axis=0))


# min value in each row
print(M.min(axis=1))
```

## Example: What is the Average Height of US Presidents?

```python
import pandas as pd

data = pd.read_csv('data/president_heights.csv')
heights = np.array(data['height(cm)'])
print(heights)
```

```python
print("Mean Height:     ", heights.mean())
print("Standard Deviation:     ", heights.std())
print("Min Height:     ", heights.min())
print("Max Height:     ", heights.max())
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn; seaborn.set() # set plot style
```

```python
plt.hist(heights)
plt.title("Height Distribution of US Presidents")
plt.xlabel("height(cm)")
plt.ylabel("number")
```
